```python
import numpy as np
import math as math
import matplotlib.pyplot as plt

```

### logs and exponential 

$$y = e^x \quad \Rightarrow \quad  y = exp(x)$$  
$$y = ln(x) \quad \Rightarrow \quad  y = log(x)$$ 
$$y = log_{10}(x) \quad \Rightarrow \quad  y = log10(x)$$ 
$$y = log_e(x) \quad \Rightarrow \quad  y = log(x)$$ 

e^x is also above zero - never gets to zero
e^x is 1 where x is 0. 

log(exp(x)) = x - log() and exp are the inverse of each other.



```python
x = np.linspace(-3,3,20)

y = math.e ** x

print(f"y is {y}")
plt.plot(x,y,'-')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.grid()
plt.show()
```

```python
x = np.linspace(0.1,2000,100)

y = np.log(x)

print(f"y is {y}")
plt.plot(x,y,'-')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.grid()
plt.show()
```

For log(x) the larger x becomes the more slowly 'y' grows. 

```python
y = np.log(np.exp(3))
print(f"y is {y}")
```

### logistic function 

$$f(x) = \frac{1}{1+e^{- \beta x}}$$

function for a sigmoid (s curve)





```python
x = np.linspace(-5,5,20)
y = 1/(1 + math.e ** -x)

plt.plot(x,y,'-')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.grid()
plt.show()
```

```python
x = np.linspace(-5,5,20)
y = 5/(1 + math.e ** -x)

plt.plot(x,y,'-')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.grid()
plt.show()
```
